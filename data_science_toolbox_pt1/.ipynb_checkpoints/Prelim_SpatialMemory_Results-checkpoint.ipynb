```python
# Set paths and import modules (Don't worry much about this)
# Import modules
import pandas as pd
import numpy as np
import os
from psychopy.misc import fromFile
import matplotlib.pyplot as plt
import fnmatch
import seaborn as sns

plt.rcParams['figure.dpi'] = 150 # default for me was 75

project_dir = '/Users/sportsnoah14/Documents/Northshore_Hosptial/Mehta-lab_Northwell/spatial_memory+sleep_task'
data_dir = 'data'
sub_id = 'NS140'

# TestPhase
postsleep_block = 'B82'
trial_path_post = project_dir + os.sep + data_dir + os.sep + sub_id + os.sep + postsleep_block
postsleepcsv = trial_path_post + os.sep + 'NS140_B82_TestPhase_WideText.csv'

presleep_block = 'B72'
trial_path_pre = project_dir + os.sep + data_dir + os.sep + sub_id + os.sep + presleep_block
presleepcsv = trial_path_pre + os.sep + 'NS140_B72_TestPhase_WideText.csv'

overnightsounds = '/Users/sportsnoah14/Desktop/spatial_memory_overnight_sounds.xlsx'

# PVT
block1 = 'B46'
trial1 = project_dir + os.sep + data_dir + os.sep + sub_id + os.sep + block1
csvpvt1 = 'NS140_B46_PVT.csv'
csv1 = trial1 + os.sep + csvpvt1

block2 = 'B69'
trial2 = project_dir + os.sep + data_dir + os.sep + sub_id + os.sep + block2
csvpvt2 = 'NS140_B69_PVT.csv'
csv2 = trial2 + os.sep + csvpvt2

block3 = 'B81'
trial3 = project_dir + os.sep + data_dir + os.sep + sub_id + os.sep + block3
csvpvt3 = 'NS140_B81_PVT.csv'
csv3 = trial3 + os.sep + csvpvt3
```

```python
# Defining functions below

######################################
# PVTPhase
######################################
def pvt_compare(**kwargs):
    
    all_blocks = pd.DataFrame(columns = ['Time'])
    
    fig, ax = plt.subplots( len(kwargs) + 1 ,1)
    a = -1
    meanpresses = {}
    for blocknum, pvtcsv in kwargs.items():
        a = a + 1
        pvtblock = pd.read_csv(pvtcsv, index_col = 0, names = ['TrialN', 'Time'])
        # Eliminate trials that are less than 100ms, false starts
        pvtblock = pvtblock[pvtblock['Time'] >= 100]
        # Histogram of push times
        ax[a].hist(pvtblock['Time'], bins = range(100,600,20))
        ax[a].set_title(str(blocknum) + ' PVT')
        all_blocks = pd.concat([ all_blocks, pvtblock], sort = False, ignore_index = True)
        meanpresses[blocknum] = pvtblock['Time'].mean()
        
    ax[a + 1].hist(all_blocks['Time'].tolist(), bins = range(100,600,20))
    ax[a + 1].set_title('All Trials Combined')
    ax[a + 1].set_xlabel('Time (ms)')
    
    fig.subplots_adjust(wspace=0.5,
        hspace=0.7,
        left=0.125,
        right=0.9,
        top=0.9,
        bottom=0.1)
    
    #fig.tight_layout()
    plt.show()
    return meanpresses

######################################
# TestPhase Comparison
######################################

def sleep_compare(presleep, postsleep, sleep_sounds, units = "cm"):
    
    if units == "cm":
        distmeasure = 'DistanceFromBingoCM'
    elif units == "pix":
        distmeasure = 'DistanceFromBingoPix'
    
    sleepsounds = pd.read_excel(sleep_sounds, usecols = [0])
    ssounds = sleepsounds['Sound ID'].tolist()

    all_scores = {}
    presleep = pd.read_csv(presleep)
    postsleep = pd.read_csv(postsleep)
    
    presleep['Phase'] = ['PreSleep'] * len(presleep)
    postsleep['Phase'] = ['PostSleep'] * len(postsleep)
    presleep['soundplayed'] = presleep.SoundID.isin(ssounds)
    postsleep['soundplayed'] = postsleep.SoundID.isin(ssounds)
    
    # Score in total for both blocks
    pre_sleep_score_raw = presleep['HitOrMoss'].sum()
    post_sleep_score_raw = postsleep['HitOrMoss'].sum()
    all_scores['PresleepTotal'] = pre_sleep_score_raw
    all_scores['PostsleepTotal'] = post_sleep_score_raw
    
    # Filter both data frames for only sleep sounds
    presleep_sleepsounds = presleep[presleep.SoundID.isin(ssounds)]
    postsleep_sleepsounds = postsleep[postsleep.SoundID.isin(ssounds)]
    
    # Filter for sounds not played at night
    presleep_notplayed_sounds = presleep[ presleep.SoundID.isin(ssounds) ==  False ]
    postsleep_notplayed_sounds = postsleep[ postsleep.SoundID.isin(ssounds) == False ]
    
    # Scores for sounds played during sleep
    presleep_score_sounds = presleep_sleepsounds['HitOrMoss'].sum()
    postsleep_score_sounds = postsleep_sleepsounds['HitOrMoss'].sum()
    all_scores['PresleepSoundsPlayedAtSleep'] = presleep_score_sounds
    all_scores['PostsleepSoundsPlayedAtSleep'] = postsleep_score_sounds
    
    
    # Calculate average distance from target
    # Presleep mean and sd
    presleep_distanceoff_sounds_mean = presleep_sleepsounds[distmeasure].mean()
    presleep_distanceoff_sounds_sd = presleep_sleepsounds[distmeasure].std()
    all_scores['PresleepDistanceOffMeanSoundsPlayedAtSleep'] = presleep_distanceoff_sounds_mean
    all_scores['PresleepDistanceOffSDSoundsPlayedAtSleep'] = presleep_distanceoff_sounds_sd
    
    # Postsleep mean and sd
    postsleep_distanceoff_sounds_mean = postsleep_sleepsounds[distmeasure].mean()
    postsleep_distanceoff_sounds_sd = postsleep_sleepsounds[distmeasure].std()
    all_scores['PostsleepDistanceOffMeanSoundsPlayedAtSleep'] = postsleep_distanceoff_sounds_mean
    all_scores['PostsleepDistanceOffSDSoundsPlayedAtSleep'] = postsleep_distanceoff_sounds_sd
    
    # Presleep Distance off only those incorrect
    presleep_distanceoff_ssounds_incorrect = presleep_sleepsounds[presleep_sleepsounds['HitOrMoss'] == 0]
    all_scores['PresleepDistanceOffSoundsPlayedAtSleepIncorrect'] = presleep_distanceoff_ssounds_incorrect[distmeasure].mean()
    
    # Postsleep Distance off only those incorrect
    postsleep_distanceoff_ssounds_incorrect = postsleep_sleepsounds[postsleep_sleepsounds['HitOrMoss'] == 0]
    all_scores['PostsleepDistanceOffSoundsPlayedAtSleepIncorrect'] = postsleep_distanceoff_ssounds_incorrect[distmeasure].mean()

    
    # Examine all sounds in boxplot
    combined_data = pd.concat([presleep, postsleep], ignore_index = True, sort = False)
    #h = sns.boxplot(x = "Phase", y = distmeasure, hue = 'HitOrMoss', data = combined_data)
    combined_data['soundplayed'] = combined_data['soundplayed'].replace(True, 'Sound +').replace(False, 'Sound -')
    #combined_data = combined_data.replace(False, 'Sound -')
    
    
    # Examine only sounds played at night in boxplot
    combined_data_sleepsounds = pd.concat([presleep_sleepsounds, postsleep_sleepsounds], ignore_index = True, sort = False)
    #j = sns.boxplot(x = "Phase", y = distmeasure, hue = 'HitOrMoss', data = combined_data_sleepsounds)
    
    #sns.lineplot(x = 'Phase', y = distmeasure, hue = "soundplayed", data = combined_data)
    #sns.pointplot(x = 'Phase', y = distmeasure, hue = "soundplayed", 
    #              units = "SoundID", data = combined_data)
    
    #sns.lineplot(x = 'Phase', y = distmeasure, hue = "SoundID", 
     #             style = "soundplayed", data = combined_data)
    
    g = sns.catplot(x="Phase", y = distmeasure,
            hue="SoundID", col="soundplayed",
            data=combined_data, kind="point",
            dodge=True,
            height=4, aspect=.7);

    g.set_titles("{col_name}")

    g._legend.remove()
    plt.show()
    
    return all_scores, g #,h, j
```
